# **Lasso Regression For Feature Selection**

```python
# prompt: final_test_ is my test data and final_train_ is my train data fit lasso for feature selection on this data from scratch give r2, radj,rmse,mape value for train and test

import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split


X_train = final_train.drop('TotalExpense', axis=1)
y_train = final_train['TotalExpense']
X_test = final_test.drop('TotalExpense', axis=1)
y_test = final_test['TotalExpense']

# Create and train the Lasso model (you can adjust the alpha parameter)
lasso_model = Lasso(alpha=0.1)  # You can experiment with different alpha values
lasso_model.fit(X_train, y_train)

# Make predictions on the training and test sets
y_train_pred = lasso_model.predict(X_train)
y_test_pred = lasso_model.predict(X_test)

# Calculate R-squared, Adjusted R-squared, RMSE, and MAPE for train and test sets
def calculate_adjusted_r2(r2, n, p):
  """Calculates the Adjusted R-squared."""
  return 1 - (1 - r2) * (n - 1) / (n - p - 1)

def calculate_mape(y_true, y_pred):
  """Calculates the Mean Absolute Percentage Error (MAPE)."""
  return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def evaluate_model(y_true, y_pred, X):
  r2 = r2_score(y_true, y_pred)
  rmse = np.sqrt(mean_squared_error(y_true, y_pred))
  n = len(y_true)
  p = X.shape[1]
  adjusted_r2 = calculate_adjusted_r2(r2, n, p)
  mape = calculate_mape(y_true, y_pred)
  return r2, adjusted_r2, rmse, mape


r2_train, adj_r2_train, rmse_train, mape_train = evaluate_model(y_train, y_train_pred, X_train)
r2_test, adj_r2_test, rmse_test, mape_test = evaluate_model(y_test, y_test_pred, X_test)

print("Train Results:")
print("R-squared:", r2_train)
print("Adjusted R-squared:", adj_r2_train)
print("RMSE:", rmse_train)
print("MAPE:", mape_train)

print("\nTest Results:")
print("R-squared:", r2_test)
print("Adjusted R-squared:", adj_r2_test)
print("RMSE:", rmse_test)
print("MAPE:", mape_test)
```

```python
# prompt: perform hyperparameter tunnig for lasso

from sklearn.model_selection import GridSearchCV

# Define the parameter grid for hyperparameter tuning
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10]}  # You can adjust the range of alpha values

# Create a Lasso model
lasso_model = Lasso()

# Create a GridSearchCV object
grid_search = GridSearchCV(lasso_model, param_grid, cv=5, scoring='neg_mean_squared_error')  # You can change the scoring metric

# Fit the grid search to your training data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)

# Get the best model
best_lasso_model = grid_search.best_estimator_

# Make predictions using the best model
y_train_pred = best_lasso_model.predict(X_train)
y_test_pred = best_lasso_model.predict(X_test)

# Evaluate the best model (using the same evaluation metrics as before)
r2_train, adj_r2_train, rmse_train, mape_train = evaluate_model(y_train, y_train_pred, X_train)
r2_test, adj_r2_test, rmse_test, mape_test = evaluate_model(y_test, y_test_pred, X_test)

print("\nTrain Results (Best Model):")
print("R-squared:", r2_train)
print("Adjusted R-squared:", adj_r2_train)
print("RMSE:", rmse_train)
print("MAPE:", mape_train)

print("\nTest Results (Best Model):")
print("R-squared:", r2_test)
print("Adjusted R-squared:", adj_r2_test)
print("RMSE:", rmse_test)
print("MAPE:", mape_test)
```

```python
# prompt: select feature with non zero value and then again fit lasso for that feature and also the regression with r2,rmse,r2adj,mape

# Get the coefficients from the best Lasso model
coefficients = best_lasso_model.coef_

# Create a list of features with non-zero coefficients
selected_features = [feature for feature, coef in zip(X_train.columns, coefficients) if coef != 0]

# Create new training and test sets with only the selected features
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

# Fit a new Lasso model with the selected features
lasso_model_selected = Lasso(alpha=grid_search.best_params_['alpha'])  # Use the best alpha from GridSearchCV
lasso_model_selected.fit(X_train_selected, y_train)

# Make predictions on the training and test sets
y_train_pred_selected = lasso_model_selected.predict(X_train_selected)
y_test_pred_selected = lasso_model_selected.predict(X_test_selected)


# Evaluate the new model
r2_train_selected, adj_r2_train_selected, rmse_train_selected, mape_train_selected = evaluate_model(y_train, y_train_pred_selected, X_train_selected)
r2_test_selected, adj_r2_test_selected, rmse_test_selected, mape_test_selected = evaluate_model(y_test, y_test_pred_selected, X_test_selected)

print("\nTrain Results (Selected Features):")
print("R-squared:", r2_train_selected)
print("Adjusted R-squared:", adj_r2_train_selected)
print("RMSE:", rmse_train_selected)
print("MAPE:", mape_train_selected)

print("\nTest Results (Selected Features):")
print("R-squared:", r2_test_selected)
print("Adjusted R-squared:", adj_r2_test_selected)
print("RMSE:", rmse_test_selected)
print("MAPE:", mape_test_selected)


# Fit a new Linear Regression model with the selected features
linear_model_selected = LinearRegression()
linear_model_selected.fit(X_train_selected, y_train)

# Make predictions on the training and test sets
y_train_pred_linear_selected = linear_model_selected.predict(X_train_selected)
y_test_pred_linear_selected = linear_model_selected.predict(X_test_selected)


# Evaluate the new linear regression model
r2_train_linear_selected, adj_r2_train_linear_selected, rmse_train_linear_selected, mape_train_linear_selected = evaluate_model(y_train, y_train_pred_linear_selected, X_train_selected)
r2_test_linear_selected, adj_r2_test_linear_selected, rmse_test_linear_selected, mape_test_linear_selected = evaluate_model(y_test, y_test_pred_linear_selected, X_test_selected)

print("\nTrain Results (Linear Regression with Selected Features):")
print("R-squared:", r2_train_linear_selected)
print("Adjusted R-squared:", adj_r2_train_linear_selected)
print("RMSE:", rmse_train_linear_selected)
print("MAPE:", mape_train_linear_selected)

print("\nTest Results (Linear Regression with Selected Features):")
print("R-squared:", r2_test_linear_selected)
print("Adjusted R-squared:", adj_r2_test_linear_selected)
print("RMSE:", rmse_test_linear_selected)
print("MAPE:", mape_test_linear_selected)
```

```python
# prompt: use all possible ml algorithm to fit the data using stacking,boosting,and all impprt libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from xgboost import XGBRegressor

# Assuming X_train_selected, X_test_selected, y_train, y_test are defined from previous feature selection steps

def evaluate_model(model, X, y):
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    return r2, rmse

# 1. Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train_selected, y_train)  # Using selected features
lr_r2_train, lr_rmse_train = evaluate_model(lr_model, X_train_selected, y_train)
lr_r2_test, lr_rmse_test = evaluate_model(lr_model, X_test_selected, y_test)

# 2. Lasso Regression
lasso_model = Lasso()
lasso_model.fit(X_train_selected, y_train)  # Using selected features
lasso_r2_train, lasso_rmse_train = evaluate_model(lasso_model, X_train_selected, y_train)
lasso_r2_test, lasso_rmse_test = evaluate_model(lasso_model, X_test_selected, y_test)

# 3. Decision Tree Regression
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train_selected, y_train)  # Using selected features
dt_r2_train, dt_rmse_train = evaluate_model(dt_model, X_train_selected, y_train)
dt_r2_test, dt_rmse_test = evaluate_model(dt_model, X_test_selected, y_test)

# 4. Random Forest Regression
rf_model = RandomForestRegressor()
rf_model.fit(X_train_selected, y_train)  # Using selected features
rf_r2_train, rf_rmse_train = evaluate_model(rf_model, X_train_selected, y_train)
rf_r2_test, rf_rmse_test = evaluate_model(rf_model, X_test_selected, y_test)

# 5. Gradient Boosting Regression
gb_model = GradientBoostingRegressor()
gb_model.fit(X_train_selected, y_train)  # Using selected features
gb_r2_train, gb_rmse_train = evaluate_model(gb_model, X_train_selected, y_train)
gb_r2_test, gb_rmse_test = evaluate_model(gb_model, X_test_selected, y_test)

# 6. XGBoost Regression
xgb_model = XGBRegressor()
xgb_model.fit(X_train_selected, y_train)  # Using selected features
xgb_r2_train, xgb_rmse_train = evaluate_model(xgb_model, X_train_selected, y_train)
xgb_r2_test, xgb_rmse_test = evaluate_model(xgb_model, X_test_selected, y_test)

# Ensemble Methods (Stacking) - Using selected features
estimators = [
    ('lr', lr_model),
    ('lasso', lasso_model),
    ('rf', rf_model),
    ('gb', gb_model),
    ('xgb', xgb_model)
]

stacking_model = StackingRegressor(estimators=estimators, final_estimator=LinearRegression())
stacking_model.fit(X_train_selected, y_train)  # Using selected features

stacking_r2_train, stacking_rmse_train = evaluate_model(stacking_model, X_train_selected, y_train)
stacking_r2_test, stacking_rmse_test = evaluate_model(stacking_model, X_test_selected, y_test)


print("Linear Regression - Train R2:", lr_r2_train, "Test R2:", lr_r2_test)
print("Lasso Regression - Train R2:", lasso_r2_train, "Test R2:", lasso_r2_test)
print("Decision Tree - Train R2:", dt_r2_train, "Test R2:", dt_r2_test)
print("Random Forest - Train R2:", rf_r2_train, "Test R2:", rf_r2_test)
print("Gradient Boosting - Train R2:", gb_r2_train, "Test R2:", gb_r2_test)
print("XGBoost - Train R2:", xgb_r2_train, "Test R2:", xgb_r2_test)
print("Stacking - Train R2:", stacking_r2_train, "Test R2:", stacking_r2_test)
```
